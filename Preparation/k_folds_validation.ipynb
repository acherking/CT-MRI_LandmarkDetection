## Check K-folds validation


```python
## Divided

# k = 5

import numpy as np

base_dir = "/Users/achs/PhD/code/CT-MRI_LandmarkDetection/Spartan/trained_models/divided_dataset/straight_model/mean_two_landmarks/5_cross"

k5_folds_err = []
for i in range(5):
    file_path = f"{base_dir}/{str(i)}/train_val_err_array.npy"
    err_array = np.load(file_path)
    k5_folds_err.append(err_array)

k5_folds_err_divided = np.asarray(k5_folds_err)
k5_folds_err_divided = k5_folds_err_divided ** (1/2)
```

```python
from matplotlib import pyplot

fig, axs = pyplot.subplots(3, 2, sharex=True, sharey=True, constrained_layout=True, figsize=(20, 10))

epochs_100 = np.arange(0, 100)

# based on the Volume array
# Cut from start to Row ascending order
axs[0][0].plot(epochs_100, k5_folds_err_divided[0, 0, :], 'r--')
axs[0][0].plot(epochs_100, k5_folds_err_divided[0, 1, :], 'b--')
axs[0][0].set_xlabel("K(1/5)-folds, epochs")
# Cut from end to Row descending order
axs[0][1].plot(epochs_100, k5_folds_err_divided[1, 0, :], 'r--')
axs[0][1].plot(epochs_100, k5_folds_err_divided[1, 1, :], 'b--')
axs[0][1].set_xlabel("K(2/5)-folds, epochs")
fig.suptitle("Stage 1: K5-folds validation, train_err & val_err", fontsize=16)
# Cut from start to Column ascending order
axs[1][0].plot(epochs_100, k5_folds_err_divided[2, 0, :], 'r--')
axs[1][0].plot(epochs_100, k5_folds_err_divided[2, 1, :], 'b--')
axs[1][0].set_xlabel("K(3/5)-folds, epochs")
# Cut from end to Column descending order
axs[1][1].plot(epochs_100, k5_folds_err_divided[3, 0, :], 'r--')
axs[1][1].plot(epochs_100, k5_folds_err_divided[3, 1, :], 'b--')
axs[1][1].set_xlabel("K(4/5)-folds, epochs")
# Cut from start to Slice ascending order
axs[2][0].plot(epochs_100, k5_folds_err_divided[4, 0, :], 'r--')
axs[2][0].plot(epochs_100, k5_folds_err_divided[4, 1, :], 'b--')
axs[2][0].set_xlabel("K(5/5)-folds, epochs")

pyplot.ylim(0.5, 5.5)
pyplot.setp(axs[:, 0], ylabel='(mm)')

pyplot.show()
```

```python
k5_folds_err_divided_mean = np.sum(k5_folds_err_divided, axis=0) / 5
```

```python
import matplotlib.pyplot as plt

epochs_100 = np.arange(0, 100)

plt.plot(epochs_100, k5_folds_err_divided_mean[0], 'r--')
plt.plot(epochs_100, k5_folds_err_divided_mean[1], 'b--')

plt.xlabel('Epochs')
plt.ylabel('K5 Mean Error Distance(mm)')
plt.show()
```
