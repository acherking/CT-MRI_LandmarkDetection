```python
## Divided

# k = 5

import numpy as np

base_dir = "/Users/achs/PhD/code/CT-MRI_LandmarkDetection/Spartan/trained_models/divided_dataset/straight_model/mean_two_landmarks/5_cross"

k5_folds_err = []
for i in range(5):
    file_path = f"{base_dir}/{str(i)}/train_val_err_array.npy"
    err_array = np.load(file_path)
    k5_folds_err.append(err_array)

k5_folds_err_divided = np.asarray(k5_folds_err)
k5_folds_err_divided = k5_folds_err_divided ** (1/2)
```
